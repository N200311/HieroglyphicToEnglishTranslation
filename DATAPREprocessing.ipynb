```python
import re
import pandas as pd
import numpy as np
import re
```

```python
# Target cleaning
def clean_traduction(text):
    # Start from double spaces and sentences to delete
    text = " ".join(text.split())
    if text.endswith("..."):
        text = text[:-3].strip()
    if text == "?":
        text = text.replace("?", "")
    if "-??-" in text:
        text = ""
    text = re.sub(r"--.*?--", "--zerst√∂rt--", text)
    if "--zerst√∂rt--" in text:
        text = ""
    if "..." in text:
        text = ""
    if "‚Ä¶" in text:
        text = ""
    if ". . ." in text:
        text = ""
    if "_" in text:
        text = ""
    if "‚∏Æ_?" in text:
        text = ""
    if "?_?" in text:
        text = ""
    if "---?---" in text:
        text = ""
    if "---" in text:
        text = ""
    if "--" in text:
        text = ""
    if "keine √úbersetzung vorhanden" in text:
        text = ""
    if "Keine √úbersetzung m√∂glich" in text:
        text = ""
    if "--- LEER GEFUNDEN ---" in text:
        text = ""
    if "---LEER GEFUNDEN---" in text:
        text = ""
    text = re.sub(r"\(=.*?\)", "", text)
    # if text == 'The':
    # text = ''
    if "[---]" in text:
        text = ""
    # lhg acronym, other languages, special parenthesis and chapter numbers
    text = re.sub(r"\(\((.*?)\)\)", r"\1", text)
    text = re.sub(r"\[\[(.*?)\]\]", r"\1", text)
    text = text.replace('"arbustes √† √©pines"', "dornige Str√§ucher")
    text = text.replace("r√¥deurs", "pl√ºnderer")
    text = re.sub('\\"(.*?)"', r"\1", text)
    text = re.sub(r"(\/[\w+√Ñ√§√ñ√∂·∫û√ü√ú√º]+)", " ", text)
    text = text.replace("- LHG -", " Leben, Heil, Gesundheit ")
    text = text.replace("- LHG", " Leben, Heil, Gesundheit ")
    text = text.replace("-LHG", " Leben, Heil, Gesundheit ")
    text = text.replace("- {LHG} LHG -", " Leben, Heil, Gesundheit ")
    text = text.replace("LHG", "Leben, Heil, Gesundheit")
    text = text.replace("l.h.g.", "Leben, Heil, Gesundheit")
    text = text.replace("l.h,.g.", "Leben, Heil, Gesundheit")
    text = text.replace("l.h-g", "Leben, Heil, Gesundheit")
    text = text.replace("l.h.g .", "Leben, Heil, Gesundheit")
    text = text.replace("l.h.g -", "Leben, Heil, Gesundheit")
    text = text.replace("l.h.g", "Leben, Heil, Gesundheit")
    text = text.replace("l.p.h.", "Life, Prosperity, Health")
    text = text.replace("LPH", "Life, Prosperity, Health")
    text = text.replace("‚Äû", "").replace("‚Äú", "").replace("‚Äù", "")
    text = text.replace("‚∏¢", "").replace("‚∏£", "")
    text = re.sub(r"\$\[.*?\]\$", "", text)
    text = text.replace("[", "").replace("]", "")
    text = text.replace("<", "").replace(">", "")
    text = text.replace("ìâò", "").replace("ìäÇ", "")
    text = text.replace("ìçπ", "").replace("ìç∫", "")
    text = text.replace("‚Äö", "").replace("‚Äò", "")
    text = re.sub(r"‚∏Æ(.*?)\?", r"\1", text)
    # text = re.sub('\((.*?)\)[^\|]', ' ', text) !Attention! Problems with other parenthesis
    text = re.sub(r"\((.*?)\)\|", r"\1", text)
    text = text.replace("|", "")
    text = re.sub(r"\[¬ß[0-9]+\]", "", text)
    text = re.sub(r"\[¬ß[0-9]+\w+\]", "", text)
    text = re.sub(r"¬ß[0-9]+(\s|\.|$|\,|\:|.*?)", r"\1", text)
    text = re.sub(r"¬ß\s[0-9]+(\s|\.|$|\,|\:|.*?)", r"\1", text)
    text = re.sub(r"¬ß\s[0-9]+-[0-9]+(\s|\.|$|\,|\:|.*?)", r"\1", text)
    text = re.sub(r"\-\s(Variante)(.*?)\-", "", text)
    text = re.sub(r"^(Variante)(.*?)$", r"\2", text)
    text = re.sub(r"(Variante)(.*?)$", "", text)
    # und, von, OA, UA acronyms and comments inside parenthesis
    text = text.replace("u.", "und")
    text = text.replace("v.", "von")
    text = text.replace(". ---", "")
    text = text.replace("--NN--", "").replace("|NN|", "").replace("NN", "")
    text = re.sub(r"\(w√∂rt.*?\)", "", text)
    text = re.sub(r"\(w√∂rt.*?$", "", text)
    text = re.sub(r"\[√§ltere Fassung.*?\]", "", text)
    text = re.sub(r"\(√§lterer Text.*?\)", "", text)
    text = re.sub(r"\(oder.*?\)", "", text)
    text = re.sub(r"^\[Beischrift.*?\]:", "", text)
    text = re.sub(r"\[Beischrift.*?\]", "", text)
    text = re.sub(r"\[.*?Beischrift.*?\]", "", text)
    text = re.sub(r"(O.?√Ñg?\.?)", "Ober√§gypten", text)
    text = re.sub(r"(U.?√Ñg?\.?)", "Unter√§gypten", text)
    text = text.strip("'").strip('"')
    text = text.strip()
    text = text.lstrip(".")
    # ‚å©‚å™ and {} parenthesis, and other elements
    text = re.sub(r"\{(.*?)\}\‚å©(.*?)\‚å™", r"\1", text)
    text = re.sub(r"\‚å©(.*?)\‚å™\{(.*?)\}", r"\2", text)
    text = text.replace("‚å©‚å©", "").replace("‚å™‚å™", "")
    text = text.replace("{{", "").replace("}}", "")
    text = re.sub(r"(\{.*?\}\s+[\w√Ñ√§√ñ√∂·∫û√ü√ú√º.,=:]+\s+)\‚å©(.*?)\‚å™", r"\1", text)
    text = re.sub(r"\‚å©(.*?)\‚å™(\s+[\w√Ñ√§√ñ√∂·∫û√ü√ú√º.,=:]+\s+\{.*?\})", r"\2", text)
    text = re.sub(
        r"(\{.*?\}[\w√Ñ√§√ñ√∂·∫û√ü√ú√º.,=:]+\s+[\w√Ñ√§√ñ√∂·∫û√ü√ú√º.,=:]+\s+)\‚å©(.*?)\‚å™", r"\1", text
    )
    text = re.sub(
        r"\‚å©(.*?)\‚å™([\w√Ñ√§√ñ√∂·∫û√ü√ú√º.,=:]+\s+[\w√Ñ√§√ñ√∂·∫û√ü√ú√º.,=:]+\s+\{.*?\})", r"\2", text
    )
    text = re.sub(
        r"(\{.*?\}\s+[\w√Ñ√§√ñ√∂·∫û√ü√ú√º.,=:]+\s+[\w√Ñ√§√ñ√∂·∫û√ü√ú√º.,=:]+\s+)\‚å©(.*?)\‚å™", r"\1", text
    )
    text = re.sub(
        r"\‚å©(.*?)\‚å™(\s+[\w√Ñ√§√ñ√∂·∫û√ü√ú√º.,=:]+\s+[\w√Ñ√§√ñ√∂·∫û√ü√ú√º.,=:]+\s+\{.*?\})", r"\2", text
    )
    text = re.sub(r"\‚å©(.*?)\‚å™(\s+[\w√Ñ√§√ñ√∂·∫û√ü√ú√º.,=:]+\{.*?\})", r"\2", text)
    text = re.sub(r"(\{.*?\}[\w√Ñ√§√ñ√∂·∫û√ü√ú√º.,=:]+\s+)\‚å©(.*?)\‚å™", r"\1", text)
    text = re.sub(r"\‚å©(.*?)\‚å™([\w√Ñ√§√ñ√∂·∫û√ü√ú√º.,=:]+\s+\{.*?\})", r"\2", text)
    text = re.sub(r"\{(.*?)\}\s\‚å©(.*?)\‚å™", r"\1", text)
    text = re.sub(r"\‚å©(.*?)\‚å™\s\{(.*?)\}", r"\2", text)
    text = re.sub(r'"(.*?)\/(.*?)"', r"\1", text)
    text = text.replace("‚å©", "").replace("‚å™", "")
    text = text.replace("{", "").replace("}", "")
    text = text.replace("Íú§", "Íú•")
    text = text.replace("`", "'")
    text = text.replace("#", "")
    text = text.replace("‚â°", "=")
    text = text.replace("&", "und")
    text = text.replace("$", "")
    text = text.replace("(?)", "")
    text = re.sub(r"\.\s(oder[\s\w√Ñ√§√ñ√∂·∫û√ü√ú√º.,=:]+)", "", text)
    text = text.replace("*", "")
    text = text.replace('"', "")
    text = re.sub(r"\(.*?\)", "", text)
    text = re.sub(r"\(d\.h\.\s[\s\w√Ñ√§√ñ√∂·∫û√ü√ú√º.,=:]+", "", text)
    # Double spaces again
    text = " ".join(text.split())
    return text
```

```python
# Step 2: Load your CSV file
input_csv = 'preprocessed_dataset.csv'
df = pd.read_csv(input_csv)
clean_df = df.copy()

# Step 3: Apply cleaning function to the 'source' column
df['target_cleaned'] = df['target'].apply(clean_traduction)

# Step 4: Save to a new CSV file
output_csv = 'PREeprocessed_dataset.csv'
df.to_csv(output_csv, index=False, encoding='utf-8')

print(f"‚úÖ Cleaning done! Cleaned data saved to {output_csv}")
```
